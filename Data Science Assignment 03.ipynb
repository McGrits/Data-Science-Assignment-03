```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://webpages.charlotte.edu/mscipion/"

r = requests.get(url)
print(r.text)
```

```python
soup = BeautifulSoup(r.text, "html")
print(soup)
```

```python
table = soup.find("table")
print(table)
```

```python
headers = table.find_all("th")
print(headers)
```

```python
titles = []

for i in headers:
    title = i.text
    titles.append(title)
    
print(titles)
```

```python
df= pd.DataFrame(columns=titles)
print(df)
```
